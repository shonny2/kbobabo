```python
import pandas as pd
import numpy as np
import seaborn as sns
```

### 울산 날씨데이터 (2022 ~ 2023)

```python
US_df22 = pd.read_csv('152_2223.csv', encoding = 'cp949')
US_df23 = pd.read_csv('152_2324.csv', encoding = 'cp949')

US_df22['지점'] = '울산'
US_df23['지점'] = '울산'

US_df = pd.concat([US_df22, US_df23], axis = 0)
US_df.head(3)
```

```python
US_df.columns

# 필요한 컬럼
# '지점', '일시', '평균기온(°C)', '최저기온(°C)', '최고기온(°C)', '일강수량(mm)', '평균 풍속(m/s)', '평균 상대습도(%)', 

ulsan_wth = US_df[['지점', '일시', '평균기온(°C)', '최저기온(°C)', '최고기온(°C)', '일강수량(mm)', '평균 풍속(m/s)', '평균 상대습도(%)']]

ulsan_wth = ulsan_wth.rename(columns={'평균기온(°C)' : '평균기온',
                                      '최저기온(°C)' : '최저기온', 
                                      '최고기온(°C)' : '최고기온', 
                                      '일강수량(mm)' : '강수량', 
                                      '평균 풍속(m/s)' : '풍속', 
                                      '평균 상대습도(%)' : '습도'})

ulsan_wth
```
