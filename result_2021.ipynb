```python
import pandas as pd
import numpy as np 
```

### 데이터 불러오기


```python
df21 = pd.read_csv('result - 2021.csv')
```

> 데이터 전처리(구단명 통일)
  - 전부 두글자로 통일
  - KIA > 기아
  - SSG -> SK

```python
# 기아 이름바꾸기

df21['원정'] = df21['원정'].str.replace("KIA", "기아", 1)
df21['홈'] = df21['홈'].str.replace("KIA", "기아", 1)

# SSG 이름바꾸기

df21['원정'] = df21['원정'].str.replace("SSG", "SK", 1)
df21['홈'] = df21['홈'].str.replace("SSG", "SK", 1)
```

```python
df21
```

> 데이터 전처리(점수)

- 원정팀    : 왼/두글자 추출
- 원정점수  : 왼/세번째부터 추출
- 홈점수    : 오/세번째부터 추출
- 홈팀      : 오/두글자 추출 

```python
def game(data) :
    data['원정팀'] = data['원정'].str[:2]
    data['A'] = data['원정'].str[2:]
    data['홈팀'] = data['홈'].str[-2:]
    data['B'] = data['홈'].str[:-2]
    
    
game(df21)
df21
```

> 데이터 전처리(게임결과)

  - 1. 빈칸일 경우 : 우천취소
  - 2. A-B >0 일 경우 : 원정팀 출력 / A-B <0 일 경우 : 홈팀 출력 / A-B =0 일 경우 : 무승부 

```python
def score(data) :    
    data['A'] = pd.to_numeric(data['A'])
    data['B'] = pd.to_numeric(data['B'])
    
    data['result'] = data['B'] - data['A']
    
    
score(df21)
```

```python
df21
```

```python
df21['승리팀'] = df21['result'].fillna('우천취소')

df21.loc[df21['result'] >0, '승리팀'] = df21['홈팀']
df21.loc[df21['result'] <0, '승리팀'] = df21['원정팀']
df21.loc[df21['result'] == 0, '승리팀'] = '무승부'

df21
```
